```python
import sqlalchemy
from sqlalchemy import create_engine, MetaData, insert, inspect, text
```

```python
engine = create_engine('sqlite:///:memory:')
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
```

```python
dados.head()
```

```python
dados.to_sql('Clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
querry = 'SELECT * FROM Clientes WHERE Categoria_de_renda="Empregado"'
```

```python
empregados = pd.read_sql(querry, engine)
```

```python
empregados
```

```python
empregados.to_sql('Empregados1', con=engine, index=False)
```

```python
pd.read_sql('Empregados1', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual' ])
```

```python
querry = 'SELECT * FROM Clientes'
```

```python
pd.read_sql(querry, engine)
```

```python
query = 'DELETE FROM Clientes WHERE ID_Cliente=5008804'
with engine.connect() as conn:
    conn.execute(text(query))
    conn.commit()
```

```python
pd.read_sql_table('Clientes', engine)
```

```python
query = 'UPDATE Clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'
with engine.connect() as conn:
    conn.execute(text(query))
    conn.commit()
```
